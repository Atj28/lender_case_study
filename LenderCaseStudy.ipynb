```python
data2['emp_title'] = data2['title'].fillna('Others')
```

```python
data2['emp_title'] = data2['emp_title'].fillna('Others')
```

```python
data2['emp_length'] = data2['emp_length'].fillna('NA')
```

```python
# Convert the 'revol_util' column to string type, forcing any non-string values to NaN
data2['revol_util'] = data2['revol_util'].astype(str)

# Remove '%' from the revol_util column and convert to float
data2['revol_util'] = data2['revol_util'].str.replace('%', '', regex=False)

# Convert to float, coercing errors (like empty strings) to NaN
data2['revol_util'] = pd.to_numeric(data2['revol_util'], errors='coerce')

# Calculate the mean of the revol_util column, ignoring NaN values
mean_revol_util = data2['revol_util'].mean()

# Replace NaN values in the revol_util column with the mean value
data2['revol_util'] = data2['revol_util'].fillna(mean_revol_util)
```

```python
data2['last_pymnt_d'] = data2['last_pymnt_d'].fillna('NA')
data2['collections_12_mths_ex_med'] = data2['collections_12_mths_ex_med'].fillna('NA')
data2['last_credit_pull_d'] = data2['last_credit_pull_d'].fillna('NA')
data2['chargeoff_within_12_mths'] = data2['chargeoff_within_12_mths'].fillna('NA')
data2['pub_rec_bankruptcies'] = data2['pub_rec_bankruptcies'].fillna('NA')
data2['tax_liens'] = data2['tax_liens'].fillna('NA')
```

```python
# Define income ranges
def categorize_income(income):
    if income <= 300000:
        return 'Up to 3 Lakh'
    elif income <= 1000000:
        return '3 to 10 Lakh'
    else:
        return '10 Lakh and above'

# Create a new column for income ranges
data['income_range'] = data['annual_inc'].apply(categorize_income)
```

```python
# Create a contingency table
income_loan_status = pd.crosstab(data['income_range'], data['loan_status'])

# Display the table
print(income_loan_status)
```

```python
import matplotlib.pyplot as plt

# Plotting the bar chart
income_loan_status.plot(kind='bar', stacked=True)
plt.title('Loan Status by Income Range')
plt.xlabel('Income Range')
plt.ylabel('Number of Loans')
plt.legend(title='Loan Status')
plt.xticks(rotation=0)
plt.show()
```

```python
# Calculate percentage of each loan status within each income range
income_loan_status_percentage = income_loan_status.div(income_loan_status.sum(axis=1), axis=0) * 100

# Display the percentages
print(income_loan_status_percentage)
```
